```python
res = requests.get('https://datalab.naver.com/keyword/realtimeList.naver', headers=headers)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
soup
```

```python
rank_list = soup.select(".ranking_list .ranking_item")
# 띄어쓰기 의미: .ranking_list 하위의 .ranking_item을 가져와
# 둘 사이에 > 가 있으면 직속만 가져올 수 있음. 현재 예시에서는 ranking_list > ranking_item > item_box
rank_list
```

```python
for rank in rank_list:
    print(rank.text)
```

```python
# 공백 기준으로 나눠볼까?
for rank in rank_list:
    print(rank.text.split())
```

```python
rank_list = soup.select(".ranking_list .item_title")
rank_list
```

```python
for (i, rank) in enumerate(rank_list):
    print(i+1, rank.text)
```

```python
rank_list = soup.select(".ranking_list .item_title")
rank_result_list = [] # 결과값을 담을 빈 리스트 생성
for (index, rank) in enumerate(rank_list):
    # 이용하기 쉽도록 Dictionary 형태로 구성
    rank_result_list += [{"rank":index+1, "value":rank.text}]

# 결과값이 의도대로 들어갔는지 출력하여 확인
print(rank_result_list)
print("-"*100)

# 마음껏 사용
print(f'실시간 급상승 검색어 {rank_result_list[0]["rank"]}위는 "{rank_result_list[0]["value"]}"입니다.')
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
driver = webdriver.Chrome('./chromedriver')
```

```python
# 페이지가 모두 로드될 때까지 3초 기다려주는 코드
driver.implicitly_wait(3)
# 셀레니움을 통해 실제 크롬창에서 접속한 것과 동일하게 해당 URL 내용 가져옴
driver.get('https://datalab.naver.com/keyword/realtimeList.naver')
# 소스를 읽어 BeautifulSoup를 이용해 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# driver.close() 할일 다 하고 창 닫기

# 그 다음은 위에서 했던 것과 동일하게 처리
rank_list = soup.select('.ranking_list .item_title')
rank_result_list = [] 
for (index, rank) in enumerate(rank_list):
    # 이용하기 쉽도록 Dictionary 형태로 구성
    rank_result_list += [{"rank":index+1, "value":rank.text}]

# 결과값이 의도대로 들어갔는지 출력하여 확인
print(rank_result_list)
print("-"*100)

# 마음껏 사용
print(f'실시간 급상승 검색어 {rank_result_list[0]["rank"]}위는 "{rank_result_list[0]["value"]}"입니다.')
```

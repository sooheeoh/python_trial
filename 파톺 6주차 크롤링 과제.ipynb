````python
tds = the_table[0].find_all('td')
회차 = tds[4].text
방영일 = tds[5].text
닐슨 = tds[6].text
TMNS = tds[7].text 

print(f"|회차: {회차} | 방영일: {방영일} | 닐슨: {닐슨} | TMNS: {TMNS}")
````

````python
def printInfo(i, 회차, 방영일, 닐슨, TMNS):
    회차 = tds[4].text
    방영일 = tds[5].text
    닐슨 = tds[6].text
    TMNS = tds[7].text
    
print(f"{방영일}에 방영한 {회차}차 프로그램의 시청률을 알려드리겠습니다. 닐슨은 {닐슨}, TMNS는 {TMNS}입니다.")

for i, scr in enumerate(the_table):
    회차 = scr.find_all('td')
    방영일 = scr.find_all('td')
    닐슨 = scr.find_all('td')
    TMNS = scr.find_all('td')
    printInfo(i, 회차, 방영일, 닐슨, TMNS)
````

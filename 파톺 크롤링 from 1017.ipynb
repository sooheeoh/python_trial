```python
import requests
from bs4 import BeautifulSoup

# requests     
res = requests.get("https://news.v.daum.net/v/20201004215700006")
print(res.content)
```

```python
# BeautifulSoup  
soup = BeautifulSoup(res.content, 'html.parser')
print(soup)
```

```python
#   
title = soup.find('title')
title
```

```python
#     
#       !
# java       . ^^
print(title.text)
```

```python
#    h2      
h2s = soup.find_all('h2')
h2s
```

```python
# get(   )  class (attribute)  
for h2 in h2s:
    print(h2.get('class'))
```

```python
import requests
from bs4 import BeautifulSoup

# requests를 이용해 원하는 웹페이지의 소스코드 불러옴
res = requests.get("https://news.v.daum.net/v/20201004215700006")
print(res.content)
```

```python
# BeautifulSoup를 이용해 파싱
soup = BeautifulSoup(res.content, 'html.parser')
print(soup)
```

```python
# 필요한 데이터를 검색
title = soup.find('title')
title
```

```python
# 필요한 데이터 값만 추출하여 출력
# 이것 탭에 있는 제목이지 웹페이지 제목이 아님!
# java에서 크롤링 하려면 몇백 줄 정도 나온다고 해요. ^^
print(title.text)
```

```python
# 해당 코드 내에서 h2라는 태그를 모두 검색하여 리스트 형태로 저장
h2s = soup.find_all('h2')
h2s
```

```python
# get(속성 값 불러오는 애)을 통해 class라는 속성(attribute)의 값을 불러옴
for h2 in h2s:
    print(h2.get('class'))
```

```python
# 개발자 도구에서 특정 부분 커서 하고 Elements에서 Copy - Copy Selector 해서 복붙
soup.select('#cSub > div > h3')
```

```python
# select는 무조건 리스트로 가져옴 그래서 이렇게
soup.select('#cSub > div > h3')[0].text
```
